```python
# Calculate cost and profit percentage
data['Cost']=data['Sales']-data['Profit']

data['Profit%']=(data['Profit']/data['Cost'])*100
```

```python
data.head()
```

# Multivariate Visual Insights

```python
# Calculate correlation matrix
datacorr = data.corr()
# Plot heatmap 
sns.heatmap(datacorr, annot=True, cmap='YlGnBu')
# Set title
plt.title('Correlation between variables')
```

Determining the most common shipping methods and identifying the preferred mode of shipping products.

```python
# Group data by 'Ship Mode'
ship_mode_types = data.groupby('Ship Mode')
for mode, group_df in ship_mode_types:
    print(mode)
```

```python
data.groupby('Ship Mode').groups
```

```python
data['Ship Mode'].value_counts()
```

```python
sns.histplot(x=data['Ship Mode'], color='b')
plt.title('Preferred mode of shipping products')
```

Segmentation of Customers

```python
segment_types = data.groupby('Segment')
for segment,df in segment_types:
    print(segment)
```

```python
data['Segment'].value_counts()
```

```python
sns.histplot(x=data['Segment'], color ='r')
plt.title('Customer Segments')
```

Category-based Analysis

```python
category = data.groupby('Category')
for cat,df in category:
    print(cat)
```

```python
sns.countplot(x=data['Category'], palette='Set2')
plt.title('Product Categories')
```

```python
sns.countplot(x=data['Region'], hue=data['Category'])
plt.title('Product Categories Ordered by Region')
```

The South orders fewer technology products and more office supplies, while the West orders the most technology products among all regions.

```python
sns.scatterplot(x=data['Ship Mode'], y=data['Sales'], hue=data['Category'])
```

This visualization helps to explore the relationship between sales and shipping modes across different product categories.

```python
ds =data.groupby('Category')['Profit','Sales'].agg('sum')
print(ds)
ds.plot.bar(color=sns.color_palette('pastel'))
plt.legend(loc='upper left')
plt.title('Profit and Sales by Category')
```

Analysis by Sub-Category

```python
subcatarr = []
subcat = data.groupby('Sub-Category')
for i,df in subcat:
    print(i)
    subcatarr.append(i)
```

```python
plt.figure(figsize=(7,7))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title('Distribution of Orders Across Sub-Categories')
plt.show()
```

Analysis by Region

```python
regions = data.groupby('Region')
for i,df in regions:
    print(i)
```

```python
rw =data.groupby('Region')['Profit','Sales'].agg('sum')
rw.plot.bar(color=sns.color_palette('pastel'))
plt.legend(loc='upper left')
plt.title('Profit and Sales Analysis by Region')
```

```python
plt.figure(figsize=(10,10))
data['Region'].value_counts().plot.pie(autopct="%1.1f%%" ,colors=sns.color_palette("pastel"))
```
